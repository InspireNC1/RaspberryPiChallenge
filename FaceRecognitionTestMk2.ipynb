```python
import face_recognition
import cv2
import pprint
import time
import csv
import sys

def saveName(Name):
    with open("/Users/akshar/anaconda3/TestAttendance.csv", 'w') as csvfile:
        filewriter = csv.writer(csvfile, delimiter=',',
                            quotechar='|', quoting=csv.QUOTE_MINIMAL)
        filewriter.writerow(['Number', 'Name', 'TimeStamp'])
        #filewriter.writerow([Name, time.ctime()])
        try:
            writer = csv.writer(f)
            while True:
                i = 1
                filewriter.writerow( (i+1, name, time.ctime()) )
                i = i + 1
        finally:
            f.close()
        cv2.putText(frame,'Thanks for Checking in',(1,500), font, 2,(255,0,0),2,cv2.LINE_AA)
    return
                

# Get a reference to webcam #0 (the default one)
video_capture = cv2.VideoCapture(0)

# Load a sample picture and learn how to recognize it.
obama_image = face_recognition.load_image_file("/Users/akshar/dlib/obama.jpeg")
pragyan_image = face_recognition.load_image_file("/Users/akshar/dlib/pragyan.JPG")
jashi_image = face_recognition.load_image_file("/Users/akshar/dlib/jashi.JPG")

obama_face_encoding = face_recognition.face_encodings(obama_image)[0]
pragyan_face_encoding = face_recognition.face_encodings(pragyan_image)[0]
jashi_face_encoding = face_recognition.face_encodings(jashi_image)[0]

# Initialize some variables
face_locations = []
face_encodings = [3]
face_names = []
process_this_frame = True
known_encodings = [3]
f = open(sys.argv[1], 'wt')

known_encodings = [
    obama_face_encoding,
    pragyan_face_encoding,
    jashi_face_encoding,
]

while True:
    # Grab a single frame of video
    ret, frame = video_capture.read()

    # Resize frame of video to 1/2 size for faster face recognition processing
    small_frame = cv2.resize(frame, (0, 0), fx=0.25, fy=0.25)

    # Only process every other frame of video to save time
    if process_this_frame:
        # Find all the faces and face encodings in the current frame of video
        face_locations = face_recognition.face_locations(small_frame)
        face_encodings = face_recognition.face_encodings(small_frame, face_locations)

        face_names = []
        for face_encoding in face_encodings:
            
            # See if the face is a match for the known face(s)
            match = face_recognition.compare_faces(known_encodings, face_encoding)
            name = "Unknown"
            if match[0]:
                name = "obama"
                
            if match[1]:
                name = "akshar"
                saveName(name)
                
            if match[2]:
                name = "jashi"
                
            face_names.append(name)

    process_this_frame = not process_this_frame
    

    #face_distance = face_recognition.face_distance(known_encodings, image_to_test_encoding)

    # Display the results
    for (top, right, bottom, left), name in zip(face_locations, face_names):
        # Scale back up face locations since the frame we detected in was scaled to 1/4 size
        top *= 4
        right *= 4
        bottom *= 4
        left *= 4

        # Draw a box around the face
        cv2.rectangle(frame, (left, top), (right, bottom), (0, 0, 255), 2)

        # Draw a label with a name below the face
        cv2.rectangle(frame, (left, bottom - 35), (right, bottom), (0, 0, 255), cv2.FILLED)
        font = cv2.FONT_HERSHEY_DUPLEX
        cv2.putText(frame, name, (left + 6, bottom - 6), font, 1.0, (255, 255, 255), 1)
        
    #cv2.rectangle(frame, (0, 0), (0, 0), (0, 0, 255), 2)
    
    # Display the resulting image
    cv2.imshow('Video', frame)

    # Hit 'q' on the keyboard to quit!
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release handle to the webcam
video_capture.release()
cv2.destroyAllWindows()
```
